```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df.columns
```

```python
df.head(10)
```

```python
plt.hist(df['SalePrice'], bins=20)
plt.xlabel('Sale Price')
plt.ylabel('Number of Sales')
plt.show()
```

```python
# Scatter Plot
plt.scatter(df['GrLivArea'], df['SalePrice'])
plt.xlabel('Sales Price')
plt.ylabel('Living Area')
plt.show()
```

```python
plt.scatter(df['LotArea'], df['SalePrice'])
plt.xlabel('Sale Price')
plt.ylabel('Basement Area')
plt.show()
```

```python
df['PSF'] = (df['SalePrice'] / df['GrLivArea']).round(2)
df
```

```python
df.shape
```

```python
plt.hist(df['PSF'], bins=20, color='green')
plt.xlabel('Price per Sq. Ft')
plt.ylabel('Number of Sales')
plt.show()
```

```python
df.boxplot(column=['PSF'], by=['YrSold'], figsize=(10, 6))
plt.suptitle('Price Per Sq Foot by Year Sold')
plt.title('')
plt.xlabel('Year Sold')
plt.ylabel('Price per Sq. Ft')
plt.show()
```

```python
df['Age'] = df['YrSold'] - df['YearBuilt']
df
```

```python
plt.scatter(df['SalePrice'], df['Age'])
plt.xlabel('Sale Price')
plt.ylabel('Age')
plt.show()
```

```python
plt.scatter(df['Age'], df['PSF'])
plt.ylabel('Price per Sq. Ft')
plt.xlabel('Age')
plt.show()
```

```python
df[df['YrSold'] == 2007].boxplot(column=['PSF'], by=['Neighborhood'], figsize=(10, 6), rot=90)
plt.suptitle('Price per Sq. Ft by Neighborhood')
plt.title('2007')
plt.xlabel('Neighborhood')
plt.ylabel('Sq. Ft')
plt.show()
```

```python
df[df['YrSold'] == 2007].boxplot(column=['PSF'], by=['BldgType'], figsize=(10, 6), rot=90)
plt.suptitle('Price per Sq. Ft by Building Type')
plt.title('2007')
plt.xlabel('Building Type')
plt.ylabel('Sq. Ft')
plt.show()
```

```python
df[df['YrSold'] == 2007].boxplot(column=['PSF'], by=['BedroomAbvGr'], figsize=(10, 6), rot=90)
plt.suptitle('Price per Sq. Ft by Above Ground Bedroom')
plt.title('2007')
plt.xlabel('Above Ground Bedroom')
plt.ylabel('Sq. Ft')
plt.show()
```

```python
df[df['YrSold'] == 2007].boxplot(column=['PSF'], by=['OverallQual'], figsize=(10, 6), rot=90)
plt.suptitle('Price per Sq. Ft by Overall Quality')
plt.title('2007')
plt.xlabel('Overall Quality')
plt.ylabel('Sq. Ft')
plt.show()
```

```python
df[df['YrSold'] == 2007].boxplot(column=['PSF'], by=['OverallCond'], figsize=(10, 6), rot=90)
plt.suptitle('Price per Sq. Ft by Overall Condition')
plt.title('2007')
plt.xlabel('Overall Condition')
plt.ylabel('Sq. Ft')
plt.show()
```

```python
df['Functional'].value_counts().plot(kind='bar')
plt.ylabel('Price Per Sq. Feet')
plt.xlabel('Functionality')
plt.show()
```

```python
df['SaleCondition'].value_counts().plot(kind='bar')
plt.ylabel('Price Per Sq. Feet')
plt.xlabel('Sale Condition')
plt.show()
```
