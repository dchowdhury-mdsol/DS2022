# Question 2 

    Using the supplied database, create helper functions that will do the following
    (The table to work with is called cleaned_data)

#### RETURN ROWS BY BOROUGH

```python
def return_by_borough(borough):
    connector = sqlite3.connect('eviction.db')
    cursor = connector.cursor()
    query = cursor.execute('select * from cleaned_data where borough=:b', {'b':borough.upper()})
    connector.close
    return [i for i in query.fetchall()]

return_by_borough('bronx')
```

#### RETURN ROWS BY NEIGHBORHOOD

```python
def return_by_neighborhood(n):
    connector = sqlite3.connect('eviction.db')
    cursor = connector.cursor()
    query = cursor.execute('select * from cleaned_data where nta=:n', {'n':n.title()})
    connector.close
    return [i for i in query.fetchall()]

return_by_neighborhood('east new york')
```

### *BONUS*  RETURN ROWS BY DATE

```python
def return_by_date(date):
    connector = sqlite3.connect('eviction.db')
    cursor = connector.cursor()
    query = cursor.execute('select * from cleaned_data where execute_date>=:d', {'d':date})
    connector.close
    return [i for i in query.fetchall()]

return_by_date(2020)
```

# Question 3

    Answer these questions about the dataset.

```python
# Converting db to dataframe
import pandas as pd

connector = sqlite3.connect('eviction.db')
cursor = connector.cursor()
query = cursor.execute('select * from cleaned_data')
data_lst = [i for i in query.fetchall()]

df = pd.DataFrame(data_lst, columns=[
    'id', 'court_index_number', 'docket_number', 
    'eviction_address', 'eviction_apt_number', 'execute_date', 
    'marshal_first_name', 'marshal_last_name', 'residential_or_commercial',
    'borough', 'eviction_postcode', 'ejectment', 'eviction_legal_possession',
    'latitude', 'longitude', 'community_board', 'council_district', 'census_tract',
    'bin', 'bbl', 'nta'
    ])
df
```
