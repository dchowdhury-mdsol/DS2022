```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('Netflix subscription fee Dec-2021.csv')
```

```python
top_10, bottom_10, random_10 = df.sort_values(by=['Total Library Size'], ascending=False), df.sort_values(by=['Total Library Size'], ascending=True), df.sample(10)
top_10, bottom_10 = top_10.nlargest(10, ['Total Library Size']), bottom_10.head(10)
top_10
```

```python
bottom_10
```

```python
random_10
```

```python
# Chart 1 - Histogram -> Distribtion of TV Shows and Movies and Total Library Size
plt.hist(df['No. of TV Shows'], label='Number of TV Shows')
plt.hist(df['No. of Movies'], color='y', label='Number of Movies')
plt.hist(df['Total Library Size'], label='Total Library Size')
plt.legend()
plt.show()
```

```python
# Chart 2 - Bar Plot -> Country vs Library Size
bar1 = plt.bar(top_10['Country'], top_10['Total Library Size'], color='magenta', label='Top 10')
bar2 = plt.bar(bottom_10['Country'], bottom_10['Total Library Size'], color='g', label='Bottom 10')
plt.xticks(rotation=90)
plt.ylabel('Total Library Size')
plt.title('Total Library Size by Country')
plt.legend()
plt.show()
```

```python
# Chart 3 - Scatter Plot -> Price vs Country
fig = plt.figure(figsize=(9.5, 3))
axs = fig.add_axes([0, 0, 1, 1])
axs.scatter(df.Country, df['Cost Per Month - Basic ($)'], color='r', marker='p', label='Cost Per Month - Basic ($)')
axs.scatter(df.Country, df['Cost Per Month - Standard ($)'], color='cyan', marker='s', label='Cost Per Month - Standard ($)')
axs.scatter(df.Country, df['Cost Per Month - Premium ($)'], color='pink', label='Cost Per Month - Premium ($)')
plt.legend(bbox_to_anchor=(1.3, 1)) # This is only from Python 3.6+ btw
plt.ylabel('Cost in USD')
plt.title('Cost by Country')
plt.xticks(rotation=90)
plt.show()
```
